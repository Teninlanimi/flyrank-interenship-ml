## 2. Discover a real truth yourself

The safest, most satisfying early wins are **things you find in the data** — un-leakable, and they *are* the core lesson. Run the three cells below. Each is ~10 lines of pandas and each overturns a common SEO belief.

Every number is **computed live from the shipped CSV** — nothing is hardcoded.

```python
import pandas as pd, numpy as np
df = pd.read_csv("data/raw/content_refresh_anonymized.csv")
print(df.shape[0], "rows,", df.shape[1], "columns")
df.head(3)
```

### Discovery A — "High search volume means more traffic." Does it?

```python
corr = df["search_volume"].corr(df["impressions_90d"])
print(f"Correlation between search_volume and impressions_90d: {corr:.3f}")
print("Near zero -> keyword search volume barely predicts the traffic a page actually gets.")
```

### Discovery B — the CTR cliff by position
Click-through rate is not flat: it collapses as you move down the results.

```python
visible = df[df["impressions_90d"] >= 100]
ctr_by_pos = visible.groupby("position_tier")["ctr"].mean().sort_values(ascending=False)
print(ctr_by_pos.round(4).to_string())
ctr_by_pos.plot(kind="bar", title="Mean CTR by position tier (impressions >= 100)", ylabel="CTR");
```

### Discovery C — is longer content the lever?
Compare word count for **declining** vs **growing** pages.

```python
wc = df.groupby("trend_direction")["word_count"].median()
print(wc.round(0).to_string())
print("\n'down' vs 'up' pages have almost the same median word count -> length is not the lever.")
```

Pick **one** of these and write a few lines below. This is your Week-1 discovery — you'll reference it in your Week-1 research-question write-up (on the InternHQ board).

**Rules:** describe what you observe as *observed / directional* — never "I proved Google's algorithm." Keep client data out of anything you publish.

```python
# Your discovery here
corr = df["search_volume"].corr((df["impressions_90d"] > 0).astype(int))

print(f"Correlation: {corr:.3f}")
print("This measures whether higher search volume is associated with pages receiving at least one impression.")
```

```python
print((df["impressions_90d"] > 0).value_counts())
```

Correlation = NaN indicates that the correlation could not be computed. In this case, the variable representing whether a page received any impressions (impressions_90d > 0) had no variation because every page in the dataset received at least one impression. Since Pearson correlation requires both variables to vary, the calculation is undefined, resulting in a NaN value.

```python
visible = df[df["impressions_90d"] >= 100]

ctr_by_pos = (
    visible.groupby(["position_tier", "content_type"])["ctr"]
    .mean()
    .sort_values(ascending=False)
)

print(ctr_by_pos.round(4))
ctr_by_pos.plot(kind="bar", title="Mean CTR by position tier (impressions >= 100)", ylabel="CTR");
```

Comparison articles have the lowest CTR in most position tiers, suggesting they generally underperform compared with feedly and keyword articles at similar ranking positions. The exception is the deep tier, where feedly articles have the lowest average CTR.

```python
# Average content age by trend direction
age_by_trend = df.groupby("trend_direction")["content_age_days"].mean().sort_values()

print(age_by_trend.round(2))

age_by_trend.plot(
    kind="bar",
    title="Average Content Age by Trend Direction",
    ylabel="Average Content Age (Days)"
)
```

If "up" has the lowest average age, newer content may be growing faster.
If "up" has the highest average age, older content may continue gaining traffic.
If the averages are very similar, content age may have little relationship with trend direction.

```python
corr = df["avg_position"].corr(df["ctr"])

print(f"Correlation between avg_position and CTR: {corr:.3f}")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))
plt.scatter(df["avg_position"], df["ctr"], alpha=0.4)
plt.xlabel("Average Position")
plt.ylabel("CTR (%)")
plt.title("Average Position vs CTR")
plt.show()
```

Typically, you would expect a negative correlation because:

Smaller position numbers (1, 2, 3) mean higher rankings.
Higher-ranked pages usually receive higher CTR.

For example:

Correlation = -0.75 → Strong negative relationship.
Correlation = -0.20 → Weak negative relationship.
Correlation ≈ 0 → Little or no linear relationship.

Do pages that rank higher in search results receive a higher click-through rate?"

```python
corr = df["avg_position"].corr(df["ctr"])

print(f"Correlation between avg_position and ctr: {corr:.3f}")

if corr < -0.5:
    print("Strong negative relationship: higher rankings are associated with higher CTR.")
elif corr < 0:
    print("Weak negative relationship: better rankings are somewhat associated with higher CTR.")
else:
    print("Little or no relationship between average position and CTR.")
```
